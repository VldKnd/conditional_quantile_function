```python
import numpy as np
import matplotlib.pyplot as plt
import torch

from datasets.synthetic.tictac import TicTacDataset
```

```python
%matplotlib inline
```

```python
tt = TicTacDataset(in_dim=5, out_dim=2, seed=13)
X, Y = tt.sample_joint(1000)
print(f"{X.shape=}, {Y.shape=}")
```

```python
fig, ax = plt.subplots(1, 1, subplot_kw={'projection': '3d'})
ax.set_xlabel("Y_0")
ax.set_ylabel("Y_1")
ax.set_zlabel("X_0")
ax.scatter(
        Y[:, 0], Y[:, 1], X[:, 2], color="blue", marker="o", s=30, alpha=0.2
    )
fig.show()
```
